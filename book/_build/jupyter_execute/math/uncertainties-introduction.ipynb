# Experimental Errors and Significant Figures

## Standard Convention

When we measure a physical quantity, we are attempting to determine its *true value*. However, in reality, it is impossible to know this true value with absolute certainty. Every measurement is subject to limitations and fluctuations stemming from our instruments, the environment, and even ourselves. This inherent lack of perfection in measurement is not a mistake in the colloquial sense, but rather a fundamental aspect of experimental science.

The term **error**, in a scientific context, refers to the difference between a measured value and the true value. Since the true value is often unknown, we use the concept of **uncertainty** to quantify the doubt or margin of error in our measurement. Uncertainty is a quantitative estimate of the error that is present in our data. Properly reporting our result with its associated uncertainty allows others to judge the quality of our experiment and to make meaningful comparisons with other results or theoretical predictions.

A complete measurement is therefore reported in two parts: the **best estimate** of the quantity and the **uncertainty** associated with that estimate.

The standard convention for reporting a measurement is to state our best estimate of the value, followed by the uncertainty. This format represents the **probable range** within which we are confident the true value lies.

In general, we can write the result of any measurement of a quantity $x$ as

$$ x = x_{\text{best}} \pm \delta x $$

where $x_{\text{best}}$ is the best estimate of a quantity $x$, $\delta x$ is called the **uncertainty**, or **error**, or **margin of error** in the measurement of $x$. This formula mean that we're confident or reasonably certain that the $x$ quantity lies somewhere between $x_{\text{best}} - \delta x$ and $x_{\text{best}} + \delta x$. However, to be more precise, we need to state a percent confidence of the uncertainty. We will return to this question after we consider the statistical laws that lead the process of measurement.

## Types of Experimental Errors

Experimental uncertainties are generated through two main types of errors: systematic error and random error. It is crucial to understand the difference between them, as they affect our results in different ways and are handled differently.

> **Systematic error** is a consistent, repeatable error that is often caused by a flaw in the experimental setup or a faulty instrument. This type of error skews measurements in a predictable and consistent direction, either always higher or always lower than the true value. Because of this consistent bias, systematic errors primarily affect the **accuracy** of a measurement, which is how close the measurement is to the true value.

**Examples of Systematic Error:**
*   A miscalibrated scale that consistently reads 0.05 grams higher than the actual mass. Every measurement taken with this scale will be biased upwards by 0.05 g.
*   Forgetting to account for the thickness of a ruler's end when measuring from the zero mark. This would cause all length measurements to be systematically shorter than they truly are.
*   An observer consistently reading a meter stick from an angle (parallax error), which would systematically shift the reading.

A key feature of systematic error is that it cannot be reduced by simply repeating the measurement multiple times and averaging the results. To reduce systematic error, we must identify its source and eliminate it by, for example, calibrating our instruments, refining our experimental technique, or applying a correction factor.

> **Random error** consists of unpredictable fluctuations in measurements. These errors arise from uncontrollable factors that change from one measurement to the next. Unlike systematic errors, random errors cause measurements to be scattered around the true value, with some being too high and some being too low. For this reason, random error primarily affects the **precision** of a measurement, which is the degree of agreement among repeated measurements of the same quantity.

**Examples of Random Error:**
*   Fluctuations in room temperature or air currents that slightly affect the reading of a sensitive balance.
*   The experimenter's reaction time when starting and stopping a stopwatch.
*   Slight variations in how an instrument is read each time.

Random errors are an inherent part of any measurement process and can never be completely eliminated. However, their effect can be minimized. The most common method to reduce the impact of random error is to take multiple measurements of the same quantity and calculate the average. The random fluctuations will tend to cancel each other out, and the average will provide a more reliable estimate of the true value.

## The Role of Significant Figures

When we report a measured value, we must be careful to only include digits that are known with some degree of reliability. 

> **Significant figures (or significant digits)** are the digits in a number that are meaningful in terms of its precision. They include all digits known with certainty, plus one final digit that is estimated or uncertain. The number of significant figures in a result implies a certain level of precision.

To correctly report and interpret measurements, we must follow a standard set of rules for determining which digits are significant:

1.  **All non-zero digits are significant.**
    * Example: 1.234 has 4 significant figures.
2.  **Zeros between non-zero digits are significant.**
    * Example: 50.081 has 5 significant figures.
3.  **Leading zeros (zeros to the left of the first non-zero digit) are not significant.** They only serve to locate the decimal point.
    * Example: 0.0075 has 2 significant figures (7 and 5).
4.  **Trailing zeros (zeros at the end of a number) are significant if the number contains a decimal point.**
    * Example: 35.00 has 4 significant figures. 7.10 has 3 significant figures. 7.100 has 4 significant figures.
5.  **Trailing zeros in a whole number without a decimal point are ambiguous.**
    * Example: The number 5200 could have 2, 3, or 4 significant figures.

To avoid this ambiguity, we use **scientific notation**.

Scientific notation removes any ambiguity regarding trailing zeros. The number is expressed as a coefficient multiplied by a power of 10. All digits in the coefficient are considered significant.
*   $5.2 \cdot 10^3$ has 2 significant figures.
*   $5.20 \cdot 10^3$ has 3 significant figures.
*   $5.200 \cdot 10^3$ has 4 significant figures.


> **Rule:** Experimental uncertainties should usually be quoted to only one or two significant figures, as more precision is unlikely to be reliable or meaningful.

**Examples**:
* $2.89 \pm 0.08$ is correct
* $2.89 \pm 0.78$ is correct
* $2.89 \pm 1.78$ is not correct  because the uncertainty has three significant figures

> **Rule:** When reporting a measurement or calculation the last significant figure in the best estimate answer should usually be the same order of magnitude (in the same decimal position) as the last significant figure of the uncertainty. This rule also has one small exception. If the last digit in the uncertainty is small (for example, 1 or, perhaps, 2), adding one extra digit in the final answer may be appropriate.

**Examples**:
* $2.80 \pm 0.08$ is correct
* $2.89 \pm 0.78$ is correct
* $2.895 \pm 0.78$ is not correct
* $85.6 \pm 3$ is not correct
* $4.7 \pm 1$ is correct because this is an exception from the aforementioned rule. One could argue that rounding it to $5 \pm 1$ would waste information.
* $85.6 \pm 2$ may be correct also as an exception.
* $85.64 \pm 2$ is not correct

**Example of the combination of both rules**. Let's say we measure a voltage and our detailed analysis gives a best estimate of $\bar{U} = 1.4827$ V and an uncertainty of $\delta U = 0.0346$ V.
1.  **Round the uncertainty:** Following our rule, we round the uncertainty to two significant figures: $\delta U \approx 0.035$ V.
2.  **Round the best estimate:** The uncertainty is in the thousandths place. Therefore, our best estimate must also be rounded to the thousandths place: $\bar{U} \approx 1.483$ V.
3.  **State the final answer:** The correct way to report this result is $U = 1.483 \pm 0.035$ V.

Stating the result as $1.4827 \pm 0.035$ V would be incorrect because the digit 7 is completely swamped by the uncertainty of 5 in the thousandths place. It is meaningless.


### Handling Significant Figures in Calculations

A common question arises when using measured values in subsequent calculations: how should significant figures be handled during intermediate steps? If you round your intermediate results to the correct number of significant figures at each step, you can introduce **rounding error**. This error can accumulate through the calculation and make your final answer less accurate than it should be.

To avoid this, it is crucial to retain at least one or two extra, non-significant digits during all intermediate calculations if possible. These are often called "[guard digits](https://en.wikipedia.org/wiki/Guard_digit)". **The best practice is to carry the full, unrounded numbers through all steps in your calculator and only apply the rules of significant figures once, at the very end, to determine the precision of your final reported answer**.

**Example: The Impact of Premature Rounding**

Suppose you need to calculate the result of $(4.58 \times 2.3) / 3.885$. The final answer should be reported to two significant figures, limited by the precision of the number 2.3.

*   **Incorrect Method (Premature Rounding):**
    1.  First, calculate the intermediate step: $4.58 \times 2.3 = 10.534$.
    2.  Rounding this intermediate result to two significant figures gives 11.
    3.  Now, perform the final step: $11 / 3.885 \approx 2.8314...$
    4.  Rounding this final result to two significant figures gives 2.8.

*   **Correct Method (Rounding Only at the End):**
    1.  First, calculate the intermediate step and keep extra digits: $4.58 \times 2.3 = 10.534$.
    2.  Use this more precise value in the final calculation: $10.534 / 3.885 \approx 2.71145...$
    3.  Rounding this final result to two significant figures gives 2.7.
