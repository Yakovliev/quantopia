# Microcanonical ensemble

## Definition

Consider a completely **isolated** physical system containing $N$ particles with total energy $E$ confined within volume $V$. Such a system operates under strict isolation conditions, meaning it cannot exchange either matter or energy with its environment.

The fundamental assumption underlying thermal equilibrium states that the probability distribution function $\rho(q, p)$ describing the system depends solely on its total energy:

$$\rho(q, p) = \rho(H(q, p)), \quad \frac{d \rho(q, p)}{dt} = 0$$

This relationship yields a time-independent $\rho(q, p)$, which aligns perfectly with ergodic theory principles and the fundamental postulate that all accessible microstates possess equal a priori probabilities

**Energy boundary layer.** 

Consider a narrow but measurable energy interval $[E, E + \Delta E]$ situated near the constant energy surface. The **microcanonical ensemble** can be mathematically expressed as:

$$
\rho(q, p) = \begin{cases}
\frac{1}{\Gamma(E, V, N)} & E < H(q, p) < E + \Delta E \\
0 & \text{elsewhere}
\end{cases}
$$ (my_first_label)

The normalization factor in this equation is given by:

$$
\Gamma(E, V, N) = \int \int_{E<H(q,p)<E+\Delta E} d^{3N}q \, d^{3N}p
$$ (my_other_label)

This represents the phase space volume encompassed by the microcanonical ensemble, specifically the region between two energy surfaces separated by $\Delta E$.

The spatial volume $V$ influences equation (8.2) through the integration boundaries for the coordinate differentials $dq$.

**Infinitesimally thin energy layer.** Define $\Phi(E, V)$ as the complete phase space volume contained within the energy surface at level $E$. This leads to:

$$\Gamma(E) = \Phi(E + \Delta E) - \Phi(E)$$

When we take the mathematical limit $\Delta E \to 0$ for an infinitesimally narrow shell, we obtain:

$$\Gamma(E) \equiv \frac{\partial \Phi(E)}{\partial E} \Delta E \equiv \Omega(E) \Delta E, \quad \Delta E \ll E \quad (8.3)$$

**State density function.** The fundamental quantity $\Omega(E)$ is defined as:

$$\Omega(E) = \frac{\partial \Phi(E)}{\partial E} = \int d^{3N}q \int d^{3N}p \, \Delta E(E - H(q, p)) \quad (8.4)$$

This represents the **density of states** at energy level $E$. Consequently, the microstate distribution $\rho$ can be written in terms of $\Omega(E)$:

$$\rho(q, p) = \begin{cases}
\frac{1}{\Omega(E)\Delta E} & E < H(q, p) < E + \Delta E \\
0 & \text{elsewhere}
\end{cases}$$

Notice the apparent mathematical singularity in $\rho(q, p)$ proportional to $1/\Delta E$. Section 8.2.4 will demonstrate that the specific value of $\Delta E$ proves to be physically irrelevant.





## Entropy

Computing expectation values for classical observables $O(q, p)$ requires averaging over the probability distribution $\rho(q, p)$ characteristic of the microcanonical ensemble:

$$\langle O \rangle = \int d^{3N}q \, d^{3N}p \, \rho(q, p) O(q, p) = \frac{1}{\Gamma(E, V, N)} \int \int_{E<H(q,p)<E+\Delta E} d^{3N}q \, d^{3N}p \, O(q, p)$$

However, **entropy** cannot be derived as a simple average of classical observables. Instead, it emerges as a function characterizing the total count of accessible quantum states.

> **FUNDAMENTAL PRINCIPLE** Following Boltzmann's insight, entropy is proportional to the natural logarithm of accessible state configurations, quantified through phase space volume $\Gamma$:
> 
> $$S = k_B \ln \left(\frac{\Gamma(E, V, N)}{\Gamma_0(N)}\right) \quad (8.5)$$

Let us now examine the implications of this fundamental definition.

**Dimensional normalization.** The normalization constant $\Gamma_0(N)$ in equation (8.5) serves dual purposes:

- $\Gamma_0(N)$ eliminates the physical dimensions of $\Gamma(E, V, N)$, ensuring the logarithmic argument remains dimensionless.

- Since particle number $N$ constitutes a fundamental thermodynamic variable, the functional relationship between $\Gamma_0(N)$ and $N$ carries significant physical importance.

**Classical mechanics limitations.** Classical statistical mechanics alone cannot uniquely determine $\Gamma_0(N)$. Subsequent analysis reveals that $\Gamma_0(N)$ assumes the value:

$$\Gamma_0(N) = h^{3N} N! \quad (8.6)$$

within quantum statistical frameworks, where $h = 6.63 \times 10^{-34} \text{m}^2\text{kg/s}$ represents Planck's constant. This quantum result is adopted for classical calculations as well.

- The factor $h^{3N}$ establishes the fundamental phase space measure with dimensions $[q]^{3N}[p]^{3N}$. Note the dimensional consistency: $[h] = \text{m (kg m/s)} = [q][p]$.

- The factorial term $N!$ accounts for quantum state counting when particles undergo permutation.

The appearance of $N!$ reflects the fundamental **indistinguishability** of identical particles.

Even when molecular motion can be accurately described using classical mechanics within certain temperature and density regimes, the essential quantum principle of particle indistinguishability cannot be ignored. Individual atomic particles cannot be tracked and labeled like distinguishable macroscopic objects.

Section 8.2.1 will explore the **Gibbs paradox**, which emerges when treating constituent particles as distinguishable entities. Under such incorrect assumptions, the factor $N!$ would be absent from $\Gamma_0(N)$.

**Entropy expressed as statistical average.** The entropy definition (8.5) can be reformulated as:

$$S = -k_B \ln \left(\frac{\Gamma_0(N)}{\Gamma(E, V, N)}\right) = -k_B \int \int d^{3N}q \, d^{3N}p \, \rho(q, p) \ln \left[\Gamma_0(N)\rho(q, p)\right] \quad (8.7)$$

This uses the fact that $\rho(q, p) = 1/\Gamma(E, V, N)$ throughout the energy shell, combined with the normalization condition:

$$\int \int d^{3N}q \, d^{3N}p \, \rho(q, p) = \int_{E<H<E+\Delta E} d^{3N}q \, d^{3N}p/\Gamma(E, V, N) = 1$$

Therefore:

$$S = -k_B \langle \ln[\Gamma_0(N)\rho(q, p)] \rangle$$

This formulation demonstrates that entropy aligns with Shannon's information-theoretic entropy definition, differing only by the multiplicative factors $k_B$ and $\Gamma_0(N)$, as discussed in Section 5.5.3.

**Thermodynamic validity requirements.** Since this entropy definition was introduced through physical reasoning rather than rigorous derivation, we must verify its consistency with established thermodynamic principles. Specifically, the entropy must satisfy:

(1) additivity property;

(2) compatibility with temperature definition;

(3) agreement with the second law of thermodynamics;

(4) adiabatic invariance principle.

### Additivity property; Gibbs paradox

Classical Hamiltonians decompose as $H(q, p) = H_{\text{kin}}(p) + H_{\text{int}}(q)$, representing kinetic energy $H_{\text{kin}}(p)$ plus inter-particle interactions $H_{\text{int}}(q)$. The constraint:

$$E < H(q, p) < E + \Delta E \quad E < H_{\text{kin}}(p) + H_{\text{int}}(q) < E + \Delta E$$

which restricts the accessible phase space volume $\Gamma(E, V, N)$ within the energy shell (equation 8.2), can be satisfied through various combinations of $H_{\text{kin}}(p)$ and $H_{\text{int}}(q)$ values.

However, the law of large numbers (discussed in Section 8.6) ensures that both kinetic and interaction energy components assume well-defined values when dealing with large particle populations $N$.

**Phase space volume scaling behavior.** Inter-particle interactions typically involve only particle pairs, while the remaining $N - 2 \approx N$ particles move relatively freely within volume $V$. This physical picture, combined with analogous reasoning for kinetic contributions, suggests that the energy shell volume scales according to:

$$\int \int_{E<H(q,p)<E+\Delta E} d^{3N}q \, d^{3N}p = \Gamma(E, V, N) \sim V^N \gamma^N(E/N, V/N) \quad (8.8)$$

Section 8.4 will verify this scaling relationship for classical ideal gas systems.

**Extensive entropy behavior.** Applying scaling relation (8.8) to the energy shell volume and utilizing $\Gamma_0(N) = h^{3N} N!$, the entropy from equation (8.5) becomes extensive:

$$S = k_B \ln \left(\frac{\Gamma(E, V, N)}{\Gamma_0(N)}\right) = k_B \ln \left(\frac{V^N \gamma^N(E/N, V/N)}{h^{3N} N!}\right)$$

$$= k_B N \left[\ln \left(\frac{V \gamma}{N h^3}\right) + 1\right] \equiv k_B N s(E/N, V/N) \quad (8.9)$$

Here we employed Stirling's approximation $N! \approx \sqrt{2\pi N}(N/e)^N$, specifically $\ln(N!) \approx N(\ln(N) - 1)$.

**Gibbs paradox resolution.** Entropy extensivity in equation (8.9) emerges from the relationship $V^N/N! \approx (V/N)^N$. Removing the factorial factor $N!$ from $\Gamma_0(N)$ would destroy extensivity, creating the **Gibbs paradox**. Classical mechanics alone cannot justify this crucial factor.

**Additivity for thermodynamically identical systems.** Two systems possess identical thermodynamic characteristics when their intensive variables match: temperature $T$, pressure $P$, particle density $N/V$, and energy density $E/N$. Equation (8.9) immediately yields:

$$S(E, V, N) = k_B (N_1 + N_2) s(E/N, V/N) = S(E_1, V_1, N_1) + S(E_2, V_2, N_2)$$

where $N = N_1 + N_2$.

**Additivity for thermally coupled systems.** Consider two systems characterized by $E_1, V_1, N_1$ and $E_2, V_2, N_2$ respectively, connected thermally to permit energy exchange while maintaining constant total energy $E = E_1 + E_2$. The entropy argument becomes:

$$\frac{\Gamma(E, V, N)}{\Gamma_0(N)} = \sum_{E_1} \frac{\Gamma(E_1, V_1, N_1)}{\Gamma_0(N_1)} \frac{\Gamma(E - E_1, V_2, N_2)}{\Gamma_0(N_2)} \quad (8.10)$$

The law of large numbers ensures the right-hand side exhibits sharp peaking at maximum value $E_1 = E_{\max}$ with peak width scaling as $\sqrt{E_{\max}}$. Consequently:

$$S(E_{\max}) < S(E, N, V) < k_B \ln \left(\sqrt{E_{\max}}\right) + S(E_{\max}) \quad (8.11)$$

The first inequality reflects that individual terms are smaller than positive sums. The second inequality in (8.11) results from replacing the sum in equation (8.10) with the product of peak width $\sqrt{E_{\max}}$ and height. Equation (8.11) defines:

$$S(E_{\max}) = k_B \ln \left(\frac{\Gamma(E_{\max}, V_1, N_1) \Gamma(E - E_{\max}, V_2, N_2)}{\Gamma_0(N_1) \Gamma_0(N_2)}\right)$$

This establishes $S(E, V, N) = S(E_1, V_1, N_1) + S(E_2, V_2, N_2)$. Note that $S(E_{\max})$ exhibits extensivity while the $\ln(E_{\max})$ term in (8.11) becomes negligible in the thermodynamic limit $N \to \infty$.

### Temperature definition compatibility

Consider two systems with entropies $S_1 = S(E_1, V_1, N_1)$ and $S_2 = S(E_2, V_2, N_2)$ in thermal contact, allowing heat exchange. At equilibrium, total entropy becomes stationary:

$$0 = dS = \frac{\partial S_1}{\partial E_1} dE_1 + \frac{\partial S_2}{\partial E_2} dE_2, \quad dE_1 = -dE_2 \quad (8.12)$$

Since total energy $E_1 + E_2$ remains constant, equilibrium condition (8.12) implies the existence of a quantity $T$ called **temperature**:

$$\frac{\partial S_1}{\partial E_1} = \frac{1}{T} = \frac{\partial S_2}{\partial E_2} \quad (8.13)$$

Temperature definition emerges directly from energy conservation. The microcanonical entropy formulation requires only that entropy depends exclusively on internal energy through the energy shell volume $\Gamma(E, V, N)$, independent of underlying microscopic dynamics.

### Second law of thermodynamics compatibility

We must confirm that statistical entropy defined by (8.5) satisfies the **second law of thermodynamics**:

> **"When an isolated system undergoes any process connecting two equilibrium states, the final state entropy cannot be less than the initial state entropy."**

**Free expansion process.** During free expansion, both energy $E$ and particle number remain constant due to the absence of external heat transfer:

$$\Delta E Q = 0$$

The energy shell volume $\Gamma(E, V, N)$ increases with expanding volume $V$. Therefore:

$$S(E, V_2, N) > S(E, V_1, N), \quad S = k_B \ln \left(\frac{\Gamma(E, V, N)}{\Gamma_0(N)}\right), \quad V_2 > V_1$$

This uses the fact that normalization factor $\Gamma_0(N)$ stays constant.

**Dynamic constraint removal.** Physical constraints (such as wall interactions) diminish when volume increases.

> **EQUIVALENCE PRINCIPLE** The second law is equivalent to stating that entropy increases when dynamic constraints are eliminated.

### Energy shell thickness considerations

Section 8.5's entropy definition involves phase space volume $\Gamma(E, V, N)$ of a shell with width $\Delta E$ centered at energy $E$. This suggests that entropy $S = S_\Delta E(E, V, N)$ depends on the unspecified parameter $\Delta E$. Does this compromise entropy's uniqueness?

**Reference energy scale.** For small $\Delta E$, the approximation becomes:

$$\Gamma(E, V, N) \approx \Omega(E) \Delta E, \quad \Omega(E) = \frac{\partial \Phi(E)}{\partial E} \quad (8.14)$$

where $\Omega(E)$ represents state density from equation (8.3). Determining whether $\Delta E$ is small or large requires a reference energy $\Delta E_0$. A natural choice is $\Delta E_0 \sim k_B T$, corresponding to typical thermal energy per particle (Section 3.4.1).

**Thermodynamic limit behavior.** Entropy involves the logarithm of $\Gamma(E, V, N)$:

$$\ln \left(\frac{\Gamma(E, V, N)}{\Gamma_0}\right) = \ln \left(\frac{\Omega(E)\Delta E \Delta E_0}{\Gamma_0 \Delta E_0}\right) = \ln \left(\frac{\Omega(E)\Delta E_0}{\Gamma_0}\right) + \ln(\Delta E/\Delta E_0)$$

ensuring dimensionless logarithmic arguments. The crucial insight is that exact values of both $\Delta E$ and $\Delta E_0$ become irrelevant in the thermodynamic limit $N \to \infty$ provided:

$$|\ln(\Delta E/\Delta E_0)| \ll N$$

Quantum energy discretization guarantees this condition in quantum statistics.

**Shell versus sphere approximation.** For large $\Delta E$, we can substitute total volume $\Phi(E)$ for the energy shell volume $\Gamma(E, V, N)$:

$$\ln (\Gamma(E, V, N)) \approx \ln (\Phi(E)) \quad (8.15)$$

neglecting normalization factor $\Gamma_0$. This approximation works because hypersphere volume and surface in $3N$ dimensions scale as $R^{3N}$ and $R^{3N-1}$ respectively (Section 8.3.1):

$$\ln (\Phi(E)) \sim \ln (R^{3N}) = 3N \ln(R)$$

$$\ln (\Gamma(E, V, N)) \sim \ln (R^{3N-1} \Delta E) = (3N - 1) \ln(R) + \ln(\Delta E)$$

confirming equation (8.15). This analysis omits detailed dimensional considerations, particularly reference energy $\Delta E_0$.

## Microcanonical ensemble calculations

Statistical physics calculations follow this systematic procedure:

1) **Hamiltonian formulation**

$$H(q, p) = H(q_1, \ldots, q_{3N}, p_1, \ldots, p_{3N}, z)$$

where $z$ represents external parameters (e.g., volume $V$). Function $H(q, p)$ encodes microscopic interaction details.

2) **Phase space determination and state density calculation**:

$$\Omega(E, V, N) = \int d^{3N}q \int d^{3N}p \, \Delta E(E - H(q, p))$$

3) **Entropy calculation from energy sphere volume $\Phi(E)$**:

$$S(E, V, N) = k_B \ln \left(\frac{\Phi(E)}{\Gamma_0}\right)$$

4) **Thermodynamic variable determination**:

$$\frac{1}{T} = \left(\frac{\partial S}{\partial E}\right)_{V,N}, \quad -\frac{\mu}{T} = \left(\frac{\partial S}{\partial N}\right)_{E,V}, \quad \frac{P}{T} = \left(\frac{\partial S}{\partial V}\right)_{E,N}$$

5) **Internal energy evaluation**:

$$U = \langle H \rangle = E(S, V, N)$$

6) **Thermodynamic potential calculation via Legendre transformations**:

$$F(T, V, N) = U - TS$$

$$H(S, P, N) = U + PV$$

$$G(T, P, N) = U + PV - TS$$

7) **Additional physical quantities** beyond thermodynamic potentials can be computed, including probability distributions for system properties like momentum/velocity distributions. For an $N$-particle system with phase space density:

$$\rho(q, p) = \rho(\vec{q}_1, \ldots, \vec{q}_N, \vec{p}_1, \ldots, \vec{p}_N)$$

the probability of finding particle $i$ with momentum $\vec{p}$ is:

$$\rho_i(\vec{p}) = \langle \Delta E(\vec{p} - \vec{p}_i) \rangle$$

$$= \int d^3q_1 \ldots d^3q_N \int d^3p_1 \ldots d^3p_N \, \rho(\vec{q}_1, \ldots, \vec{q}_N, \vec{p}_1, \ldots, \vec{p}_i, \ldots, \vec{p}_N) \Delta E(\vec{p} - \vec{p}_i)$$

### Hypersphere geometry

For future applications, we calculate the volume:

$$\Omega_n(R) = \int_{\sum_{i=1}^n x_i^2 < R^2} d^n x = R^n \Omega_n(1) \quad (8.16)$$

of an $n$-dimensional hypersphere with radius $P$.

**Spherical coordinate transformation.** The unit sphere volume $\Omega_n(1)$ appears in the Jacobian determinant when converting via:

$$d^n x = dx_1 \ldots dx_n = \Omega_n(1) nR^{n-1} dR \quad (8.17)$$

from Euclidean to spherical coordinates. This transformation applies when integrands depend only on radius $R$.

**Gaussian integration techniques.** To evaluate (8.16), we utilize the Gaussian integral identity:

$$\int_{-\infty}^{+\infty} dx e^{-x^2} = \sqrt{\pi}, \quad \int_{-\infty}^{+\infty} dx_1 \ldots \int_{-\infty}^{+\infty} dx_n e^{-(x_1^2+\ldots+x_n^2)} = \pi^{n/2}$$

This gives:

$$\pi^{n/2} = \int_0^{\infty} e^{-R^2} \Omega_n(1) nR^{n-1} dR = n\Omega_n(1) \int_0^{\infty} e^{-y} y^{(n-1)/2} \frac{dy}{2\sqrt{y}}$$

$$= \frac{n}{2}\Omega_n(1) \int_0^{\infty} e^{-y} y^{\frac{n}{2}-1} dy \quad (8.18)$$

using equation (8.17), the substitution $\sum_i x_i^2 = R^2 \equiv y$, and $2RdR = dy$.

**Gamma function application.** With the definition:

$$\Gamma(z) = \int_0^{\infty} dx x^{z-1} e^{-x}$$

equation (8.18) yields:

$$\pi^{n/2} = \frac{n}{2}\Omega_n(1)\Gamma(n/2), \quad \Omega_n(1) = \frac{\pi^{n/2}}{(n/2) \Gamma(n/2)} \quad (8.19)$$

Note that this evaluates hypersphere volume using formally dimensionless variables $x_i$. For subsequent applications:

$$\Gamma(N) = (N - 1)\Gamma(N - 1), \quad \Gamma(N) = (N - 1)! \quad (8.20)$$

when the Gamma function argument is an integer.
